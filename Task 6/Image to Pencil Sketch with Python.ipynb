## Task 6 Image to Pencil Sketch with Python

### Importing the libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Reading the image

```python
image = cv2.imread("girls-pic-with-lighting.jpg")
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

### Converting it to the gray scale

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_image,cmap='gray')
plt.show()
```

### Inverting the grayscale image (Negetive)

```python
inverted_image = 255- grey_image
```

```python
plt.imshow(inverted_image,cmap='gray')
plt.show()
```

### Blurring the inverted grayscale image

```python
blurred = cv2.GaussianBlur(inverted_image,(21,21),0)
```

```python
plt.imshow(blurred, cmap='gray')
plt.show()
```

### Pencil sketch 

```python
inverted_blur = 255 - blurred
pencilsketch = cv2.divide(grey_image,inverted_blur, scale=256.0)
```

```python
plt.imshow(pencilsketch, cmap='gray')
plt.show()
```

```python
#original image
plt.figure(figsize=(15,7))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()

# pencil sketch image
plt.figure(figsize=(15,7))
plt.imshow(pencilsketch, cmap='gray')
plt.show()
```
