### Feature Selection

```python
#feature selection based required task and importance
df_terror = df[[ 'iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'provstate', 'city', 'latitude',
               'longitude', 'location', 'attacktype1_txt', 'targtype1_txt', 'weaptype1_txt','gname', 'nkill',
                'nwound', 'motive', 'addnotes']]

df_terror.head()
```

### Feature Engineering

```python
#feature engineering (creating a new feature)
df_terror['Causuality'] = df_terror['nkill']+ df_terror['nwound']
```

Causuality = No of Kills + No of Wounds

### Renaming for clear understanding

```python
#Renaming
df_terror =df_terror.rename(columns = {'iyear' : 'Year', 'imonth' : 'Month', 'iday' : 'Day', 'country_txt' : 'Country', 'region_txt' : 'Region',
           'provstate' : 'Province/State', 'city' : 'City', 'latitude' : 'Latitude', 'longitude' : 'Longitude',
           'location' : 'Location', 'attacktype1_txt' : 'Attack_Type', 'targtype1_txt' : 'Target_Type',
           'weaptype1_txt' : 'Weapon_Type','gname' : 'Group_name', 'nkill' : 'Killed', 'nwound' : 'Wounded', 'motive' : 'Motive',
           'addnotes' : 'Add_Notes'} )

df_terror.head()
```

```python
df_terror.shape
```

```python
df_terror.info()
df_terror.describe()
```

#### Observations:-

1. The data consists of terrorism activities ranging from 1970 to 2017

### Number of attacks in each year

```python
# visualization for no of attacks in each year
Attacks = df_terror['Year'].value_counts().sort_index().to_frame().reset_index().rename(columns = {'index': 'Year',
                                                                                                   'Year': 'Attacks'}
                                                                                       ).set_index('Year')

print(Attacks.head())
print(Attacks.tail())
```

```python
Attacks.plot(kind= 'bar', figsize= (15,8), fontsize= 12)
plt.title('Timeline of Attacks',fontsize=15)
plt.xlabel('Year', fontsize= 14)
plt.ylabel('Number of Attacks', fontsize=14)
plt.show()
```

1). The Maximum number of attacks is 16903 in the year 2014

2). The Minimum number of attacks is 471 in the year 1971

```python
# missing value treatment for killed,wounded and causualities
df_terror['Killed'] = df_terror['Killed'].fillna(0)
df_terror['Wounded'] = df_terror['Wounded'].fillna(0)
df_terror['Causuality'] = df_terror['Causuality'].fillna(0)
```

### Number of Causualities in each year

```python
# total no of causualities(killed + wounded) in each year
Num_Causuality = df_terror[['Year','Causuality']].groupby('Year').sum()
print(Num_Causuality.head())
print(Num_Causuality.tail())
```

```python
Num_Causuality.plot(kind= 'bar', figsize= (15,8))
plt.title('Causualities in each year', fontsize=15)
plt.xlabel('Year',fontsize= 14)
plt.ylabel('Total No of Causualities', fontsize= 14)
plt.show()
```

##### 1). The least Number of Causualities are in the year 1971. Infact there were no people killed or wounded

##### 2). The Most Number of Causuality are in the year 2015. There were more than 70000 Causualities

### Country Analysis

### Number of attacks in each country

```python
# No of attacks in each country
Attacks_country = df_terror['Country'].value_counts().head(10)
Attacks_country
```

### Country wise causalities

```python
#country wise causalities
Country_wise_causuality =df_terror[['Country', 'Causuality']].groupby('Country').sum().sort_values(by= 'Causuality',
                                                                                                   ascending= False)
Country_wise_causuality.head(10)
```

###  Country wise killed and wounded

```python
# country wise wounded
country_wise_wounded = df_terror[['Country', 'Wounded']].groupby('Country').sum().sort_values('Wounded', ascending=False)
country_wise_wounded.head(10)
```
